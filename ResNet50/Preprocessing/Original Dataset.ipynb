```python
import numpy as np
import glob
import time
import cv2
import matplotlib.pyplot as plt
```

```python
def dataset(path = '../../Dataset'):
    '''
    Import dataset. (Need to import numpy, glob, time, cv2)
    
    Arguments:
    path -- the path of image set
    
    Returns:
    list_data -- return a numpy list that contains all of the images and their labels
    '''
    
    all_data = glob.glob(path + '/256_ObjectCategories/*')
    all_data.sort()
    
    list_data = []
    class_num = len(all_data) - 1
    for i in range(class_num):
        data_eve_class = glob.glob(all_data[i] + '/*')
        tic = time.time()
        for j in range(len(data_eve_class)):
            img = cv2.imread(data_eve_class[j])
            list_data.append([i, img])
        toc = time.time()
        print('Import NO. ' + str(i + 1) + ' class use ' + str((toc-tic)*1000) + 'ms')
            
    return list_data
```

```python
def label_dict(path = '../../Dataset'):
    '''
    To build a dict to record the label
    (need to import glob package)
    Arguments:
    path --
    
    Returns:
    dict_l --
    '''
    all_label = glob.glob(path + '/256_ObjectCategories/*')
    all_label.sort()
    
    dict_l = {}
    class_num = len(all_label) - 1
    for i in range(class_num):
        dict_l[i] = all_label[i].split('.')[-1]
    
    return dict_l
```

```python
data_set = dataset()
```

```python
dict_label = label_dict()
```

```python
print('Total number of datasets: ' + str(len(data_set)))
print('Total number of classes: ' + str(len(dict_label)))
```
