```python
import os

def create_directory_dictionary(root_directory):
    directory_dict = {}

    for dirpath, dirnames, filenames in os.walk(root_directory):
        relative_path = os.path.relpath(dirpath, root_directory)
        
        # Filter out non-.jpg files
        jpg_files = [file for file in filenames if file.lower().endswith('.jpg')]
        
        directory_dict[relative_path] = jpg_files

    return directory_dict

def find_empty_directories(directory_dict):
    empty_directories = [key for key, value in directory_dict.items() if not value]
    return empty_directories

def remove_empty_directories(directory_dict, empty_directories):
    for directory in empty_directories:
        directory_dict.pop(directory, None)

def total_number_of_files(directory_dict):
    total_files = sum(len(files) for files in directory_dict.values())
    return total_files

# Example usage:
root_directory = "Data/256_ObjectCategories"
result_dict = create_directory_dictionary(root_directory)

# Find empty directories
empty_directories = find_empty_directories(result_dict)

# Remove empty directories from the dictionary
remove_empty_directories(result_dict, empty_directories)

# Calculate the total number of files
total_files = total_number_of_files(result_dict)

# Print the updated result and total number of files
print("Updated Directory Dictionary:")
for key, value in result_dict.items():
    print(f"Subdirectory: {key}")
    print(f"Files: {value}")
    print("-" * 30)
```

```python
print(f"Total Number of .jpg Files: {total_files}")
```

```python
import matplotlib.pyplot as plt

def visualize_files_per_class(directory_dict):
    classes = list(directory_dict.keys())
    file_counts = [len(files) for files in directory_dict.values()]

    plt.figure(figsize=(10, 6))
    plt.bar(classes, file_counts, color='blue')
    plt.xlabel('Class')
    plt.ylabel('Number of Files')
    plt.title('Number of Files per Class')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

visualize_files_per_class(result_dict)
```
